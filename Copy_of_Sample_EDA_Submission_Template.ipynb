### Dataset Loading

````python
# Load Dataset
import os
DATASET_FILE = '/content/drive/MyDrive/Colab Notebooks/Teen_Mental_Health_Dataset.csv'

try:
    # Use the DATASET_FILE variable defined in the initial setup
    df = pd.read_csv(DATASET_FILE)
    print(f"Dataset '{DATASET_FILE}' loaded successfully.")
except FileNotFoundError:
    print(f"Error: Dataset file '{DATASET_FILE}' not found. Please upload the file or specify the correct path.")
    # Create an empty DataFrame or handle the error appropriately
    df = pd.DataFrame()
    print("An empty DataFrame has been created. Please ensure your CSV is uploaded and DATASET_FILE variable is correctly set.")
except Exception as e:
    print(f"An error occurred while loading the dataset: {e}")
    df = pd.DataFrame()
````

### Dataset First View

````python
# Dataset First Look
if not df.empty:
    print("\n--- First 5 rows of the dataset ---")
    display(df.head())
    print("\n--- Last 5 rows of the dataset ---")
    display(df.tail())
else:
    print("DataFrame is empty, cannot display head/tail. Please load the dataset first.")
````

### Dataset Rows & Columns count

````python
# Dataset Rows & Columns count
if not df.empty:
    print(f"Number of rows: {df.shape[0]}")
    print(f"Number of columns: {df.shape[1]}")
else:
    print("DataFrame is empty, cannot display row/column count. Please load the dataset first.")
````

### Dataset Information

````python
# Dataset Info
if not df.empty:
    print("\n--- Dataset Information ---")
    df.info()
else:
    print("DataFrame is empty, cannot display info. Please load the dataset first.")
````

#### Duplicate Values

````python
# Dataset Duplicate Value Count
if not df.empty:
    duplicate_rows = df.duplicated().sum()
    print(f"Number of duplicate rows: {duplicate_rows}")
    if duplicate_rows > 0:
        print("Consider removing duplicate rows if they are exact duplicates across all columns and not meaningful records.")
else:
    print("DataFrame is empty, cannot check for duplicate values. Please load the dataset first.")
````

#### Missing Values/Null Values

````python
# Missing Values/Null Values Count
if not df.empty:
    missing_values = df.isnull().sum()
    missing_percentage = (df.isnull().sum() / len(df)) * 100
    missing_df = pd.DataFrame({
        'Missing Count': missing_values,
        'Missing Percentage': missing_percentage
    }).sort_values(by='Missing Count', ascending=False)
    print("\n--- Missing Values Count and Percentage ---")
    display(missing_df[missing_df['Missing Count'] > 0])
    if missing_df['Missing Count'].sum() == 0:
        print("No missing values found in the dataset.")
else:
    print("DataFrame is empty, cannot check for missing values. Please load the dataset first.")
````

````python
# Visualizing the missing values
if not df.empty and df.isnull().sum().sum() > 0:
    plt.figure(figsize=(12, 6))
    sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
    plt.title('Missing Values Heatmap')
    plt.show()
elif not df.empty and df.isnull().sum().sum() == 0:
    print("No missing values to visualize.")
else:
    print("DataFrame is empty, cannot visualize missing values. Please load the dataset first.")
````
